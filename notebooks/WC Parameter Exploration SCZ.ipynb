```python
try:
    import matplotlib.pyplot as plt
except ImportError:
    import sys
    !{sys.executable} -m pip install matplotlib
    import matplotlib.pyplot as plt
    
import numpy as np

from neurolib.models.wc import WCModel
from neurolib.utils.parameterSpace import ParameterSpace
from neurolib.optimize.exploration import BoxSearch

# a nice color map
plt.rcParams['image.cmap'] = 'plasma'
```

```python
wc = WCModel()
```

```python
wc.params['exc_ext'] = 1.25
wc.params['c_excinh'] = 15.0
wc.params['c_inhexc'] = 12.0
wc.params['c_inhinh'] = 0.0
wc.params['duration'] = 2000

#parameters = ParameterSpace({"exc_ext":  np.linspace(0.7, 1.6111111111111112, 10), "c_excinh": np.linspace(9, 15, 10), "c_inhexc":  np.linspace(9, 10, 5), "c_inhinh": np.linspace(0, 3, 3)})
parameters = ParameterSpace({"c_excinh": np.linspace(8, 10.333333333333334, 20), "c_inhexc":  np.linspace(8, 9.666666666666666, 10)})

# info: chose np.linspace(0, 3, 21) or more, values here are low for testing
search = BoxSearch(wc, parameters, filename="scz_model_detailed.hdf")
```

```python
search.run()
```

```python
search.loadResults()
print("Number of results: {}".format(len(search.results)))
```

```python
for i in search.dfResults.index:
    search.dfResults.loc[i, 'max_exc'] = np.max(search.results[i]['exc'][:, -int(1000/wc.params['dt']):])
```

```python
import neurolib.utils.functions as func

frs_array = []
powers_array = []

for i in search.dfResults.index:
    tmp = search.dfResults.loc[i, 'exc']
    frs, powers = func.getPowerSpectrum(
        tmp[0,:],
        wc.params["dt"],
    )
    search.dfResults.loc[i, 'max_power'] = max(powers)
    index_of_maximum = np.where(powers == max(powers))
    search.dfResults.loc[i, 'freq_max_power'] = frs[index_of_maximum]
```

```python
plt.imshow(search.dfResults.pivot_table(values='max_exc', index = 'c_excinh', columns='c_inhexc'), \
           extent = [min(search.dfResults.c_excinh), max(search.dfResults.c_excinh),
                     min(search.dfResults.c_inhexc), max(search.dfResults.c_inhexc)], origin='lower')
plt.colorbar(label='Maximum Excitatory Activity')
plt.xlabel("Input to c_excinh")
plt.ylabel("Input to exc_ext")
plt.savefig('max_exc_detailed.png', dpi=500)
```

```python
plt.imshow(search.dfResults.pivot_table(values='freq_max_power', index = 'c_excinh', columns='c_inhexc'), \
           extent = [min(search.dfResults.c_excinh), max(search.dfResults.c_excinh),
                     min(search.dfResults.c_inhexc), max(search.dfResults.c_inhexc)], origin='lower')
plt.colorbar(label='Frequency of Maximum Power')
plt.xlabel("Input to c_excinh")
plt.ylabel("Input to exc_ext")
plt.savefig('freq_max_power_detailed.png', dpi=500)
```

```python
plt.imshow(search.dfResults.pivot_table(values='max_power', index = 'c_excinh', columns='c_inhexc'), \
           extent = [min(search.dfResults.c_excinh), max(search.dfResults.c_excinh),
                     min(search.dfResults.c_inhexc), max(search.dfResults.c_inhexc)], origin='lower')
plt.colorbar(label='Maximum Power')
plt.xlabel("Input to c_excinh")
plt.ylabel("Input to exc_ext")
plt.savefig('max_power_detailed.png', dpi=500)
```
